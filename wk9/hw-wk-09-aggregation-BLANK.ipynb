```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sandeepmj/datasets/main/ppp_excerpt.csv')
df
```

```python
## Run the code that resets the default display here to round the numbers, remove scientific notation, etc.
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
## Run the code here to return all summary statistics in one go
df.describe()
```

```python
## What is the sum of all the forgiven loans
#print(df.columns)
f"{df['ForgivenessAmount'].sum():,.2f}"
```

```python
## What is the sum of all the forgiven loans
## format with dollar sign and thousands separators
## round to zero decimals
f"${df['ForgivenessAmount'].sum():,.0f}"
```

```python
## store in variable called total_loan_forgiven
total_loan_forgiven = df['ForgivenessAmount'].sum()
total_loan_forgiven
```

```python
## return the median for CurrentApprovalAmount loans, InitialApprovalAmount loans and ForgivenessAmount loan amounts
print(df['CurrentApprovalAmount'].median())
print(df['InitialApprovalAmount'].median())
print(df['ForgivenessAmount'].median())
```

```python
##  Return the above as a frame
cols = ['CurrentApprovalAmount', 'InitialApprovalAmount', 'ForgivenessAmount']
df_median = df[cols].median().to_frame()
df_median.columns = ['median']
df_median
```

```python
## get the count, mean and median for CurrentApprovalAmount
print(df['InitialApprovalAmount'].count())
print(df['InitialApprovalAmount'].mean())
print(df['InitialApprovalAmount'].median())
```

```python
## automatically return a frame (without using to_frame()) for the count, mean and median for InitialApprovalAmount, CurrentApprovalAmount
col_list = ['InitialApprovalAmount', 'CurrentApprovalAmount']
df_auto = df[col_list].agg(['count', 'mean', 'median'])
df_auto
```

#### 1. How many medals were handed out in winter games v. summer game (return both in absolutes and as a percentage)?

#### 2. Top 10 count of gold medals by country

#### 3. Of these top 10, what percent of the gold medals do they have?

```python
#1
df_s = pd.read_csv('https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-summer.csv')
df_w = pd.read_csv('https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-winter.csv')
```

```python
df_s.columns == df_w.columns
```

```python
df_s = df_s.dropna(subset=['Medal'])
```

```python
print(df_s['Medal'].count())
print(df_s['Medal'].value_counts(normalize=True))
```

```python
df_w = df_s.dropna(subset=['Medal'])
```

```python
print(df_w['Medal'].count())
print(df_w['Medal'].value_counts(normalize=True))
```

```python
df_fin = pd.concat([df_s, df_w], ignore_index=True)
df_fin = df_fin.dropna(subset=['Medal'])
df_fin
```

```python
#2
df_gold = df_fin[df_fin['Medal'] == 'Gold'] 
df_gold
```

```python
print(df_gold['Team'].value_counts().head(10))
```

```python
#3
print(df_gold['Team'].value_counts(normalize=True).head(10)*100)
```
